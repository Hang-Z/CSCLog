```python
# case - 正常样本相关性强度-柱状图
normal = [0.28, 0.79, 0.67]
anomaly = [0.82, 0.23, 0.11]
#x = ["subsequence 0 and 1", "subsequence 0 and 1", "subsequence 0 and 1"]
x = np.arange(3)
total_width, n = 0.8, 3 # 有多少个类型，只需更改n即可，比如这里我们对比了四个，那么就把n设成4
width = total_width / n
x = x - (total_width - width) / 5
plt.bar(x, normal, color='k', width=width,label='Normal')
plt.bar(x + width, anomaly, color='grey', width=width,label='Anomaly')
plt.legend(loc = "best")
plt.ylim(0,1)
plt.ylabel("Correlation weight",size=16)
plt.xticks([0,1,2],["0, 1", "1, 2", "0, 2"])
plt.yticks(fontproperties='Times New Roman', size=14)
plt.xticks(fontproperties='Times New Roman', size=14)
plt.legend(loc="best", frameon=False)#图例
# plt.show()
plt.savefig('save/correlation.pdf', format='pdf', dpi=500, bbox_inches = 'tight')
```
